```python
data.drop('id',inplace=True, axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
lb =LabelEncoder()
data['diagnosis'] = lb.fit_transform(data['diagnosis'])
data
```

```python
x = data.iloc[:,2:-1]
x
```

```python
y = data['diagnosis']
y
```

```python
from sklearn.preprocessing import StandardScaler
st = StandardScaler()
x1 = st.fit_transform(x)
x1
```

```python
from sklearn.model_selection import train_test_split
x1_train,x1_test,y_train,y_test = train_test_split(x1,y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
reg = LogisticRegression()
```

```python
reg.fit(x1_train,y_train)
```

```python
reg.predict(x1_test)
```

```python
reg.score(x1_test,y_test)
```

```python
from sklearn.svm import SVC
clf = SVC(kernel='linear')
clf.fit(x1_train,y_train)
```

```python
clf.predict(x1_test)
```

```python
clf.score(x1_test,y_test)
```

```python
from sklearn.naive_bayes import GaussianNB
cf = GaussianNB()
```

```python
cf.fit(x1_train,y_train)
```

```python
cf.predict(x1_test)
```

```python
cf.score(x1_test,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=100,criterion='entropy')
```

```python
rfc.fit(x1_train,y_train)
```

```python
predicted_val =rfc.predict(x1_test)
predicted_val
```

```python
rfc.score(x1_test,y_test)
```
